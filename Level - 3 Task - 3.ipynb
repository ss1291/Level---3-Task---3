# Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Read the XLSB file

```python
df = pd.read_excel('C:/Users/admin/Desktop/Cognifyz Internship Projects/dataset.xlsb', engine='pyxlsb')
df
```

```python
df.head(3)
```

# Convert 'Has Online delivery' and 'Has Table booking' columns to numeric (assuming 'Yes' is 1 and 'No' is 0)

```python
df['Has Online delivery'] = df['Has Online delivery'].map({'Yes': 1, 'No': 0})
df['Has Table booking'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})
```

# Group the dataset by price range and calculate the percentage of restaurants offering online delivery and table booking

```python
price_range_delivery = df.groupby('Price range')['Has Online delivery'].mean() * 100
price_range_table_booking = df.groupby('Price range')['Has Table booking'].mean() * 100
```

# Visualize the relationship between price range and the availability of online delivery and table booking

```python
plt.figure(figsize=(10, 6))

plt.plot(price_range_delivery.index, price_range_delivery, marker='o', label='Online Delivery', color='skyblue')
plt.plot(price_range_table_booking.index, price_range_table_booking, marker='o', label='Table Booking', color='orange')

plt.title('Price Range vs. Availability of Online Delivery and Table Booking')
plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants Offering Service')
plt.xticks(price_range_delivery.index)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
